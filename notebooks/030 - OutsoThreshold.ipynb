```python
import cv2
import matplotlib.pyplot as plt 
```

```python
img = cv2.imread(r"C:\Users\aliay\OneDrive\Desktop\py_test\computer vision\dataset\otsu.png")
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
# Global Threshold
ret1,th1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
# Otsu's threshold
ret2,th2 = cv2.threshold(img,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
# Otsu's threshold After applying gussian filter 
blur = cv2.GaussianBlur(img,(3,3),0)
ret3,th3 = cv2.threshold(blur,127,255,cv2.THRESH_BINARY)
```

```python
plt.figure(figsize=(16,10))
plt.subplot(131),plt.imshow(th1,'gray'),plt.title("Global Threshold")
plt.subplot(132),plt.imshow(th2,"gray"),plt.title("Otsu Threshold")
plt.subplot(133),plt.imshow(th3,"gray"),plt.title("Otsu Threshold after blur")
```
