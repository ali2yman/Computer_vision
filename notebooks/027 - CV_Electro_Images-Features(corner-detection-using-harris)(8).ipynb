```python
import numpy as np 
import matplotlib.pyplot as plt 
import cv2
```

```python
img = cv2.imread("Pattern.png",0)
plt.imshow(img,"gray")
```

```python
# img: The input image on which you want to perform corner detection.
# 4: The size of the neighborhood considered for corner detection.
# 3: The aperture parameter of the Sobel derivative used.
# 0.04: The Harris detector free parameter. This is a constant value in the Harris 
corners = cv2.cornerHarris(img,4,3,0.04)
plt.imshow(corners,"gray")
```

```python
corners = cv2.cornerHarris(img,4,3,0.04)
corners = cv2.dilate(corners,None,iterations=3)  #None>> no selected kernal-size
plt.figure(figsize=(16,10))
plt.subplot(121),plt.imshow(img,"gray")
plt.subplot(122),plt.imshow(corners,"gray")
```

## <B>Harris Pros and cons<B> 

Harris Corner Detection, a popular algorithm in computer vision, has its own set of advantages and limitations. Understanding these can help in deciding whether it is suitable for a particular application. Here are the pros and cons of the Harris Corner Detection algorithm:

<B>Pros:<B>
    
Robustness: Harris Corner Detection is relatively robust against changes in scale, rotation, and illumination. It can reliably detect corners even in images with varying lighting conditions.

Localization: It is effective in precisely localizing corners in an image. This accuracy makes it useful for various tasks, including feature matching and tracking.

Simple Implementation: The algorithm has a relatively simple implementation, making it easy to understand and use for feature detection in computer vision applications.

Invariance to Noise: Harris Corner Detection can handle noisy images relatively well, as it involves a smoothing step using a Gaussian filter, which helps in reducing the impact of noise on the corner detection process.

<B>Cons:<B>
    
Computationally Intensive: Harris Corner Detection can be computationally expensive, particularly when dealing with large images. This can limit its real-time performance in certain applications.

Sensitivity to Parameter Selection: The performance of the algorithm can be sensitive to the selection of parameters such as the window size and the empirically determined constant 'k'. Suboptimal parameter values might result in poor corner detection.

Limited to Corners: The algorithm is primarily designed for corner detection and might not perform well in scenarios where other types of features, such as edges or blobs, need to be detected.

Performance on Textured Regions: Harris Corner Detection might produce false positives in textured regions, as it tends to detect corners in these areas as well. This can be a limitation when working with images containing a significant amount of texture.

Understanding these pros and cons can help in making informed decisions while choosing the appropriate feature detection technique for specific computer vision tasks. While the Harris Corner Detection algorithm is widely used, its limitations suggest that it might not be the best choice for every scenario.
