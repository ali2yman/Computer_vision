```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
img = np.zeros((500,500),np.uint8)
cv2.rectangle(img,(0,250),(500,500),(255),-1)
cv2.rectangle(img,(0,125),(250,250),(127),-1)

plt.hist(img.ravel(),255,[0,255])
plt.show()

plt.imshow(img,'gray')
```
