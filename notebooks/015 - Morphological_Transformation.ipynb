```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
img = cv2.imread(r'C:\Users\aliay\OneDrive\Desktop\py_test\computer vision\dataset\Morphological_Transformations\smarties2.png',0)
_,mask = cv2.threshold(img,220,255,cv2.THRESH_BINARY_INV)

kernal = np.ones((5,5),np.uint8)
dilation = cv2.dilate(mask,kernal,iterations=2)
erosion= cv2.erode(mask,kernal,iterations=2)
opening = cv2.morphologyEx(mask,cv2.MORPH_OPEN,kernal,iterations=2)
closing = cv2.morphologyEx(mask,cv2.MORPH_CLOSE,kernal,iterations=2)


cv2.imshow('gray.img',img)
cv2.imshow("mask",mask)
cv2.imshow("dilation",dilation)
cv2.imshow("erosion",erosion)
cv2.imshow("opening",opening)
cv2.imshow("closing",closing)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img  = cv2.imread(r"C:\Users\aliay\OneDrive\Desktop\py_test\computer vision\dataset\Morphological_Transformations\MorphologicalTransformation.png",0)
ret,th = cv2.threshold(img,190,255,cv2.THRESH_BINARY)

plt.imshow(th,"gray")
```

<B>Erosion</B>

It removes pixels on object boundaries and shrinks the size of objects. It is used to separate connected objects.



```python
kernel = np.ones((5,5))
erosion = cv2.erode(th,kernel,iterations=2)
plt.imshow(erosion,"gray")
```

<B>Dilation</B>

It adds pixels to the boundaries of objects in an image, increasing the size of the objects. It is used to merge nearby objects.

```python
kernel = np.ones((5,5))
dilation = cv2.dilate(th,kernel,iterations=4)
plt.imshow(dilation,"gray")
```

```python
kernel = np.ones((2,2))

img  = cv2.imread(r"C:\Users\aliay\OneDrive\Desktop\py_test\computer vision\dataset\Morphological_Transformations\finger-noise.png",0)
ret,th = cv2.threshold(img,190,255,cv2.THRESH_BINARY)

plt.imshow(th,"gray")
```

<b>Opening</b>


It is an erosion followed by dilation. It is useful in removing noise from an image.

```python
opening = cv2.morphologyEx(th,cv2.MORPH_OPEN,kernel)
plt.imshow(opening,"gray")
```

<B>Closing</B>

It is a dilation followed by erosion. It is useful in closing small holes or gaps in an object.



```python
closing = cv2.morphologyEx(th,cv2.MORPH_CLOSE,kernel)

plt.imshow(closing,"gray")
```

<B>Morphological gradient
</B>

is another operation in morphological image processing. It is a useful tool for edge detection, which essentially highlights the boundaries of objects in an image. The morphological gradient is the difference between the dilation and the erosion of an image. Mathematically, it can be represented as:

Morphological Gradient=Dilation(f)Erosion(f)

<B>"hit or miss" transformation </B>

is a morphological operation used primarily in image processing. This operation is used for pattern matching and is particularly effective in detecting specific configurations of pixels in binary images.

The idea behind the "hit or miss" transformation is to define two structuring elements: one that represents the "hits" or the target pattern and another that represents the "misses" or the background pattern. By applying this transformation, you can detect occurrences of the target pattern in the image.

The "hit or miss" transformation is useful for tasks like object detection, shape recognition, and image analysis where detecting specific configurations or patterns in binary images is essential.
