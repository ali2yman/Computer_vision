```python
import numpy as np 
import cv2
import matplotlib.pyplot as plt 
```

```python
img = cv2.imread("saltAndPepper.jpeg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

<B>1. Average Filter (Mean Filter):<b>
    
How it Works: The average filter works by replacing each pixel value in an image with the average value of its neighboring pixels. It involves sliding a window over the image and calculating the average value of the pixel intensities within that window.

<b>Pros:<B>

Effective in reducing noise, especially random variations caused by sensor or transmission issues.
    
Simple to implement and computationally efficient.
    
    
<b>Cons:<b>

Smoothing can lead to loss of fine image details and blurring of edges.
    
Less effective for images with non-uniform noise distribution.

```python
kernal = np.ones((5,5),np.float32)/25
# In the cv2.filter2D function, the -1 parameter represents the desired depth of the destination image. This value indicates that the depth of the output image should be the same as the input image.
dst = cv2.filter2D(img,-1,kernal)
```

```python
plt.figure(figsize=(16,12))
plt.subplot(121),plt.imshow(img),plt.title("Original")
plt.subplot(122),plt.imshow(dst),plt.title("blurred by conv")
```

```python
# or we can do it by this way

dst = cv2.blur(img,(5,5))
plt.figure(figsize=(16,12))
plt.subplot(121),plt.imshow(img),plt.title("Original")
plt.subplot(122),plt.imshow(dst),plt.title("box filtering")
```

<B>2. Gaussian Filter:<b>
    
How it Works: The Gaussian filter is a weighted average filter where the weights are generated using a Gaussian distribution. It smooths images by averaging each pixel with its neighboring pixels, with the weights decreasing as the spatial distance from the center pixel increases.

<B>Pros:<b>

Effectively reduces noise while preserving edges and fine details better than simple averaging.

The filter is isotropic, meaning it treats all directions equally.


<b>Cons:<B>

More computationally intensive compared to the average filter.

May not be as effective with certain types of noise, such as impulsive noise.

```python
dst = cv2.GaussianBlur(img,(29,29),0)
plt.figure(figsize=(16,12))
plt.subplot(121),plt.imshow(img),plt.title("Original")
plt.subplot(122),plt.imshow(dst),plt.title("Gaussian filtering")
```

```python
dst = cv2.GaussianBlur(img,(11,11),0)
plt.figure(figsize=(16,12))
plt.subplot(121),plt.imshow(img),plt.title("Original")
plt.subplot(122),plt.imshow(dst),plt.title("Gaussian filtering")
```

In the cv2.GaussianBlur function, the parameter 0 that you see represents the standard deviation in the X direction of the Gaussian kernel.

When the standard deviation is set to 0, OpenCV automatically calculates it based on the kernel size. If you want to set a specific standard deviation, you can replace 0 with the desired value.

The GaussianBlur function in OpenCV is typically used for image blurring or smoothing. It convolves the image with a Gaussian kernel to achieve the blurring effect. The standard deviation determines the amount of blur applied to the image. A higher standard deviation value results in more blurring, while a lower value leads to less blurring.

By setting the standard deviation to 0, you are allowing OpenCV to automatically calculate an appropriate value based on the kernel size. This can be a convenient option when you are not concerned with a specific blurring strength and want OpenCV to handle the calculation for you.

<B>3. Median Filter:<b>
    
How it Works: The median filter replaces each pixel's value with the median value of the intensities within the defined neighborhood. It is particularly useful in removing salt-and-pepper noise, where random pixels are set to the maximum or minimum intensity values.
    

<b>Pros:<b>

Efficient in removing( impulse - salt and paper )noise without significantly blurring edges.
    
Preserves details and edges better than the mean filter.
    
    
<b>Cons:<b>

Less effective for other types of noise, such as Gaussian noise.
    
Slower compared to mean filters due to the sorting process involved.

```python
dst = cv2.medianBlur(img,3)
plt.figure(figsize=(16,12))
plt.subplot(121),plt.imshow(img),plt.title("Original")
plt.subplot(122),plt.imshow(dst),plt.title("median filtering")
```
