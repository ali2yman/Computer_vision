```python
import cv2
```

```python
cap = cv2.VideoCapture(0)   # 0 >> record from the main camera 
print(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
print(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
# this is control the values of width and hight frame
# 3 >>> for width 
# 4 >>> for hight
cap.set(3,2000)
cap.set(4,2000)
# fourcc >>> 3shan yzbt el video codec w responsible for video compersion bardo 
fourcc = cv2.VideoWriter_fourcc(*'XVID')
# 20 >> 20 frame per sec   ,,,,,     (-,-)>>   width , hight
out = cv2.VideoWriter('output.avi',fourcc,20.0,(640,480))
```

```python
print(cap.isOpened())
while (cap.isOpened()):
    ret,frame = cap.read()
    if ret == True:
        print(cap.get(3))
        print(cap.get(4))
# to save the video 
        out.write(frame)

        gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
        cv2.imshow('frame',gray)

        if cv2.waitKey(1) == ord('q'):
            break
    else:
        break

cap.release
out.release
cv2.destroyAllWindows()
```
