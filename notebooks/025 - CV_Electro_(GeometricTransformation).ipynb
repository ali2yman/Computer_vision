```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
img  = cv2.imread("1879542-39587833-2560-1440.jpg")
img  = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
h = img.shape[0]
w = img.shape[1]
```

<b>Translation<b>

```python
M = np.float32([[1,0,400],
               [0,1,200]])

dst = cv2.warpAffine(img,M,(w,h))
plt.imshow(dst)
```

```python
# rotation 
M = cv2.getRotationMatrix2D((w/2,h/2),90,1)   # 3shan a3ml rotation mazbot around the center >> (90 > ylf 90 darga,, 1>Scale)

dst2 = cv2.warpAffine(img,M,(w,h))
plt.imshow(dst2)
```

```python
pts1 = np.float32([[10,100],[200,50],[100,250]])
pts2 = np.float32([[110,100],[300,50],[200,250]])

M = cv2.getAffineTransform(pts1,pts2)
dst = cv2.warpAffine(img,M,(w,h))
plt.imshow(dst)
```
