```python
import numpy as np
import cv2
import os 
os.chdir(r"C:\Users\aliay\OneDrive\Desktop\py_test\computer vision\dataset\34 -35 detect-track and count")
from tracker import *
```

```python
# Create an object from tracker 
tracker = EuclideanDistTracker()

cap = cv2.VideoCapture("highway.mp4")

# object detection from stable(not moving) camera 
# hna bnfsl el backgroung 3n elforeground  >>> el back ground byb2a black w elhaga ely fyha motion white
# history: This parameter defines the number of frames used for building the background model. In your case, it's set to 100, meaning the algorithm will consider the last 100 frames to establish the background.
# varThreshold : parameter controls how sensitive the background subtraction model is to changes in pixel values. A higher value, like 50 in your case, makes the model less responsive to small changes, reducing false positives (detecting changes that aren't really there). However, this also makes the model slower to adapt to new objects entering the scene. So, it's a trade-off between sensitivity to changes and adaptability to new objects.
object_detector = cv2.createBackgroundSubtractorMOG2(history=100,varThreshold=50)
```

```python
while True:
    ret,frame = cap.read()
    # print(frame.shape)
    height,width,_ = frame.shape

    # Extract Region of interest ely hnsht8l 3lyh bmGa en el hight 720 we elwidth 1280
    roi = frame[340:720,500:800]


    # 1 - object detection 
    mask = object_detector.apply(roi)   #da elgoz2 ely hnsht8l 3lyh (mask)
    # we apply this second threshold becouse the result from the first threshold there was a shadow appers for the objects we need to remove it
    _,mask = cv2.threshold(mask,254,255,cv2.THRESH_BINARY)

    # Use cv2.CHAIN_APPROX_SIMPLE if you want a simplified representation of the contour, where straight segments are approximated by their end points.
    # Use cv2.CHAIN_APPROX_NONE if you want to retain all the contour points without any approximation.
    contours,_ = cv2.findContours(mask,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)

    detections = []
    for cnt in contours:
        # calculate area and remove small elements(small area)
        area = cv2.contourArea(cnt)
        if area > 100 :
             # bngeeb el coordinates bta3 el recatangle 
            x,y,w,h = cv2.boundingRect(cnt)
            detections.append([x,y,w,h])


    #  2 - object tracking
    boxes_ids = tracker.update(detections)  
    for boxe_id in boxes_ids:
        x,y,w,h,id = boxe_id
        cv2.putText(roi,str(id),(x,y),cv2.FONT_HERSHEY_PLAIN,2,(255,0,0),2)
        cv2.rectangle(roi,(x,y),(x+w,y+h),(0,255,0),3)

    cv2.imshow("roi",roi)
    cv2.imshow("frame",frame)
    cv2.imshow("mask",mask)

    key = cv2.waitKey(30)
    if key == 27:
        break

cap.release()
cv2.destroyAllWindows()
```
