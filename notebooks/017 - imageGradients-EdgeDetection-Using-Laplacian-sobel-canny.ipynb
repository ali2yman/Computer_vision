<B>Laplacian Operator:<b>
    

It computes the second derivative of the image, which highlights regions of rapid intensity change. The Laplacian operator is more sensitive to noise and requires careful tuning or additional processing to produce desirable results. 

doc >> https://docs.opencv.org/3.4/d5/db5/tutorial_laplace_operator.html


```python
img = cv2.imread("messi5.jpg",0)

laplacian = cv2.Laplacian(img, -1, ksize=3)
plt.imshow(laplacian, cmap='gray')
plt.show()
```

<b>Sobel<b>
    
    It is a simple differentiation operator that computes an approximation of the gradient of the image intensity function. It uses two 3x3 kernels to detect changes in the horizontal and vertical directions. The Sobel operator is efficient and commonly used for general-purpose edge detection. 

```python
img = cv2.imread("sudoko.jpeg",0)

# -1 >>> to keep the image as it is ,, specifies that the output image will have the same depth as the input image.
# In this case, 1 indicates that we are computing the first-order derivative in the x-direction and 0 means we are not computing any derivative in the y-direction.
# ksize >> kernel size 
sobel_x = cv2.Sobel(img,-1,1,0,ksize=3)
sobel_y = cv2.Sobel(img,-1,0,1,ksize=3)
# sobel_d captures diagonal edge information.
sobel_d = cv2.Sobel(img,-1,1,1,ksize=3)
# combine_sobel_x_and_y combines the edge information from both horizontal and vertical directions.
combine_sobel_x_and_y = cv2.bitwise_or(sobel_x,sobel_y)


plt.figure(figsize=(16,10))

plt.subplot(221), plt.imshow(sobel_x, cmap="gray"), plt.title("sobel_x")
plt.subplot(222), plt.imshow(sobel_y, cmap="gray"), plt.title("sobel_y")
plt.subplot(223), plt.imshow(sobel_d, cmap="gray"), plt.title("sobel_d")
plt.subplot(224), plt.imshow(combine_sobel_x_and_y, cmap="gray"), plt.title("combine_sobel_x_and_y")

plt.show()
```

<b>scharr<B>

It is an improvement over the Sobel operator, offering better rotational symmetry and approximation of the gradient. The Scharr operator's kernels are designed to provide more accurate results for diagonal edges, typically at 45° and 135°.

```python
img = cv2.imread("car.jpeg",0)

sobel_x = cv2.Scharr(img,-1,1,0,5)
sobel_y = cv2.Scharr(img,-1,0,1,5)

plt.figure(figsize=(16,10))

plt.subplot(121), plt.imshow(sobel_x,"gray"), plt.title("Scharr_X")
plt.subplot(122), plt.imshow(sobel_y,"gray"), plt.title("Scharr_y")
plt.show()
```

<b>Berwat</b>


It is a modification of the Sobel operator, using a larger 5x5 filter mask to capture more information from the image neighborhood. The Berwat operator aims to provide improved edge detection results compared to the traditional Sobel operator by considering a broader context

<B>canny Edge Detector:<B>

It is a multi-stage algorithm known for its superior performance and accurate edge localization. The Canny edge detector includes stages for Gaussian smoothing, gradient computation, non-maximum suppression, and hysteresis thresholding for edge tracking. It is robust to noise and can detect a wide range of edges with good accuracy. 

doc >> https://docs.opencv.org/4.x/da/d22/tutorial_py_canny.html

```python
img = cv2.imread("messi5.jpg",0)

# 150: This parameter is the lower threshold value, which defines the minimum intensity gradient to be considered as an edge. Any gradient value below this threshold is considered as not an edge.
# 200: This parameter is the upper threshold value, defining the maximum intensity gradient to be considered an edge. Any gradient value above this threshold is considered as a strong edge.
# and if any connected edge from 150 to 200 with anthor above 200 >> adge
canny = cv2.Canny(img,150,200)
plt.imshow(canny, 'gray')
```
